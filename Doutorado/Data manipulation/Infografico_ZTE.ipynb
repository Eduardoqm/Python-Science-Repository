```python
#Importa o panda,o  pyplot do matplotlib e o numpy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Chama e da nome a planilha
zte=pd.read_excel("C:/Users/Eduardo Q Marques/Documents/My Jobs/Artigos/Redefinindo os limites/Infograma/Dados_ZTE.xlsx")
zte
```

```python
#Grafico 1984 km²
plt.bar(zte['Classe'], zte['1984km'], color = ['darkgreen','gray','pink','red'])
plt.ylabel('Área em km²')
plt.xticks(rotation='10')
plt.savefig('1984km',dpi=300)
```
